# Top posts using a hashtags in 24 hours

```python
import requests

# --------- CONFIGURATION ---------
ACCESS_TOKEN = long_access_token  # Facebook user access token
USER_ID = ig_user_id              # User ID of the token owner
FIELDS = "id,caption,media_type,media_url,permalink,like_count,comments_count,timestamp"

# --------- FUNCTION 1: Get Hashtag ID ---------
def get_hashtag_id(hashtag_query):
    """
    Fetch the Instagram Hashtag ID for a given hashtag name.
    """
    search_url = "https://graph.facebook.com/v23.0/ig_hashtag_search"
    search_params = {
        "user_id": USER_ID,
        "q": hashtag_query,
        "access_token": ACCESS_TOKEN
    }

    response = requests.get(search_url, params=search_params)
    if response.status_code == 200:
        data = response.json()
        hashtag_id = data['data'][0]['id']
        print(f"Hashtag ID for #{hashtag_query}: {hashtag_id}")
        return hashtag_id
    else:
        print("Error fetching hashtag ID:", response.status_code, response.text)
        return None

# --------- FUNCTION 2: Get Top Media for Hashtag ---------
def get_top_media_for_hashtag(hashtag_id):
    """
    Fetch top media posts for a given hashtag ID.
    """
    media_url = f"https://graph.facebook.com/v20.0/{hashtag_id}/top_media"
    media_params = {
        "user_id": USER_ID,
        "fields": FIELDS,
        "access_token": ACCESS_TOKEN
    }

    response = requests.get(media_url, params=media_params)
    if response.status_code == 200:
        media_data = response.json()
        for media in media_data.get("data", []):
            print("ID:", media.get("id"))
            print("Type:", media.get("media_type"))
            print("Caption:", media.get("caption"))
            print("Media URL:", media.get("media_url"))
            print("Permalink:", media.get("permalink"))
            print("Likes:", media.get("like_count"))
            print("Comments:", media.get("comments_count"))
            print("Timestamp:", media.get("timestamp"))
            print("-" * 40)

        # Pagination info
        paging = media_data.get("paging", {})
        if "next" in paging:
            print("More results available at:", paging["next"])
    else:
        print("Error fetching media:", response.status_code, response.text)


# --------- MAIN SCRIPT ---------
if __name__ == "__main__":
    hashtag_query = "millie"  # Replace with your desired hashtag
    hashtag_id = get_hashtag_id(hashtag_query)
    if hashtag_id:
        get_top_media_for_hashtag(hashtag_id)
```

## Converts permalinks to Username

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import re

def get_instagram_username_from_divs(permalink_url):
    """
    Extracts the username from Instagram post even if login/signup popup appears.
    Uses div traversal and alt attribute of the profile image.
    """
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
    driver.get(permalink_url)

    try:
        # Wait until a profile image inside the popup is visible
        profile_img = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located(
                (By.XPATH, "//div[@role='dialog']//img[contains(@alt, 'profile picture')]")
            )
        )
        alt_text = profile_img.get_attribute("alt")
        # Instagram alt text for profile picture is usually "USERNAME's profile picture"
        match = re.search(r"(\S+)'s profile picture", alt_text)
        if match:
            username = match.group(1)
            return username
        else:
            return None
    except Exception as e:
        print(f"Error extracting username: {e}")
        return None
    finally:
        driver.quit()

# --------- Example Usage ---------
post_url = "https://www.instagram.com/p/DNqwAxdS8QJ/"  # Replace with a real post URL
username = get_instagram_username_from_divs(post_url)

if username:
    print(f"The username is: {username}")
else:
    print("Could not retrieve the username.")
```
